# Virus Total API

```python
import requests
import configparser
from threading import Semaphore, Timer  

parser = configparser.ConfigParser()
parser.read('credential.txt')
API_cred = parser.get("credential","API")

hash_list = []
with open('hash-list.txt','r') as file_hash:
    hash_list = file_hash.readlines()
semaphore = Semaphore(4) # Block exceeding requests
count = 0

for line in hash_list:
    semaphore.acquire()
    try:
        count += 1
        url = "https://www.virustotal.com/api/v3/files/" + str(line)

        headers = {
            "Accept": "application/json",
            "x-apikey": str(API_cred)
        }
        response = requests.request("GET", url, headers=headers)
        result = response.text
        print(response.text)
        with open("file_out.txt", "a") as file_out:
            file_out.write(response.text)


    finally:
        timer = Timer(60, semaphore.release) # Delay with complex function
        timer.setDaemon(True)   # allows the timer to be canceled on exit
        timer.start()
```
